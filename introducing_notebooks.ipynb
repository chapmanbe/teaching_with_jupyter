# Teaching with Jupyter Notebooks

## What is Jupyter?

>Project Jupyter aims to create an ecosystem of open source tools for interactive computation and data analysis, where the direct participation of humans in the computational loop—executing code to understand a problem and iteratively refine their approach—is the primary consideration. (["The state of Jupyter"](https://www.oreilly.com/ideas/the-state-of-jupyter))

Jupyter is the evoluation from the IPython notebook to a language-independent notebook for ["literate computing"](http://blog.fperez.org/2013/04/literate-computing-and-computational.html).

>The name is inspired by Julia, Python, and R (the three open languages of data science) but represents the general ideas that go beyond any specific language: computation, data, and the human activities of understanding, sharing, and collaborating. (["The state of Jupyter"](https://www.oreilly.com/ideas/the-state-of-jupyter))
